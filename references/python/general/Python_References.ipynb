```
# part of filter visualization code
def plot_filters_single_channel_big(t):
    
    #setting the rows and columns
    nrows = t.shape[0]*t.shape[2]
    ncols = t.shape[1]*t.shape[3]
    
    
    npimg = np.array(t.numpy(), np.float32)
    npimg = npimg.transpose((0, 2, 1, 3))
    npimg = npimg.ravel().reshape(nrows, ncols)
    
    npimg = npimg.T
    
    fig, ax = plt.subplots(figsize=(ncols/10, nrows/200))    
    imgplot = sns.heatmap(npimg, xticklabels=False, yticklabels=False, cmap='gray', ax=ax, cbar=False)
```

```
def plot_filters_single_channel(t):
    
    #kernels depth * number of kernels
    nplots = t.shape[0]*t.shape[1]
    ncols = 12
    
    nrows = 1 + nplots//ncols
    #convert tensor to numpy image
    npimg = np.array(t.numpy(), np.float32)
    
    count = 0
    fig = plt.figure(figsize=(ncols, nrows))
    
    #looping through all the kernels in each channel
    for i in range(t.shape[0]):
        for j in range(t.shape[1]):
            count += 1
            ax1 = fig.add_subplot(nrows, ncols, count)
            npimg = np.array(t[i, j].numpy(), np.float32)
            npimg = (npimg - np.mean(npimg)) / np.std(npimg)
            npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
            ax1.imshow(npimg)
            ax1.set_title(str(i) + ',' + str(j))
            ax1.axis('off')
            ax1.set_xticklabels([])
            ax1.set_yticklabels([])
   
    plt.tight_layout()
    plt.show()
```

```
# data normalization

# numpy images greyscale shape (# images, height, width)
def normalize_stats_image_by_image(images):
  normalized_img = (images - images.mean(axis = (1,2), keepdims=True)) / images.std(axis = (1,2), keepdims=True)
  return normalized_img

# torch images shape (# images, channels, height, width)
def normalize_stats_image_by_image(images):
  mean = torch.mean(images,(2,3), keepdims=True)
  stdev = torch.std(images,(2,3), keepdims=True)
  return (images - mean) / stdev
```

```
def batch(batch_size, model_data, model_labels):
  """
  Creates a batch of inputs examples and the corresponding
  true labels.
  
  Args:
    batch_size: size of batches (int)
    model_data: tensor, entire dataset of inputs (num_examples, num_features)
    model_labels: tensor, entire dataset of labels (num_examples), range 0 to num_classes - 1
  
  Returns:
    input_batch: A tensor with shape (batch_size, num_features)
    label_batch: A tensor with shape (batch_size)
  """
  random_ind = np.random.choice(model_labels.size()[0], size=batch_size, replace=False)
  input_batch = model_data[random_ind]
  label_batch = model_labels[random_ind]
  return input_batch, label_batch
```

```
def val():
  """
  returns the validation loss and error rate
  """
  model.eval()

  batch_size = 100
  input_batch, label_batch = batch(batch_size, test_data, test_labels)
  output_batch = model(input_batch)
  
  # output_batch_max returns (tensor(values), tensor(indices))
  output_batch_max = torch.max(output_batch, 1)
  output_batch_values = output_batch_max[0].view(-1,1)
  output_batch_indices = output_batch_max[1]

  # computes loss
  loss = nn.CrossEntropyLoss()
  loss = loss(output_batch, label_batch)  
  
  # computes error rate by comparing predicted class and true class for all batch examples
  error_rate = 1.0 - (output_batch_indices == label_batch).float().mean()
  
  return loss.item(), error_rate.item()
```

```
# train step template

def train_step(batch_size=100):
  
  model.train()
  
  # Get a random training batch and compute corresponding outputs
  # from our model.
  input_batch, label_batch = batch(batch_size, train_data, train_labels)
  output_batch = model(input_batch)

  # output_batch_max returns (tensor(values), tensor(indices))
  output_batch_max = torch.max(output_batch, 1)
  output_batch_values = output_batch_max[0].view(-1,1)
  output_batch_indices = output_batch_max[1]

  # Compute loss (and error rate).
  loss = CrossEntropyLoss()
  loss = loss(output_batch, label_batch)  
  error_rate = 1.0 - (output_batch_indices == label_batch).float().mean()
  
  # Here's the magic of backpropagation: calling backward fills
  # in the .grad attribute for all model parameters.
  optimizer.zero_grad()
  loss.backward()
  
  # Take a step.
  optimizer.step()
  
  return loss.item(), error_rate.item()
```

```
# simple convolution neural network
# model definition
class CNN_2a(nn.Module):

    def __init__(self):
        super(CNN_2a, self).__init__()
        
        # convolution layers
        # 1 input channel, 8 output channels
        self.conv1 = nn.Conv2d(1, 8, 3, padding=1, stride=1)
        # output dimension (W - F + 2P)/S + 1, (W - F + 2P)/S + 1, D  28 x 28 x 8
        
        # 8 input channels, 16 output channels
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1, stride=1)
        # output dimension 14 x 14 x 16
        
        # fully connected layers, linear 
        self.fc1 = nn.Linear(16 * 7 * 7, 100)  # 6*6 from image dimension
        self.fc2 = nn.Linear(100, 100)
        self.fc3 = nn.Linear(100, num_classes)

    def forward(self, x):
        # Max pooling over a (2, 2) window
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        # Max pooling over a (2, 2) window
        x = F.max_pool2d(F.relu(self.conv2(x)), (2, 2))
        x = x.view(-1, self.num_flat_features(x))
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
    
      
    # function to flatten out input coming from convolution layers
    def num_flat_features(self, x):
      """
      function to flatten out input coming from convolution layers    
      function num_flat_features was taken from the pytorch beginners tutorial
      if I didn't use this I would have hard coded the size to be the dimensions of
      the output layer after the 2nd max pool which was 16 x 7 x 7
      
      Args:
        x = relevant output layer results
      """
        size = x.size()[1:]  # all dimensions except the batch dimension
        num_features = 1
        for s in size:
            num_features *= s
        return num_features
```

```
# step coding and train val loss error rate plotting

# model initialized to CNN_2a
model = CNN_2a()

# optimizer set up with learning rate
optimizer = torch.optim.SGD(model.parameters(), lr=.1)

# resets parameters every training session
for module in model.children():
  module.reset_parameters()
  

# iteration plotting taken from deep learning lab hw4
info = []
fig, ax = plt.subplots(2, 1, sharex=True)
num_steps = 5000
num_steps_per_val = 50
# resets val error to report when model reaches a new best performance
best_val_err = 1.0

for step in range(num_steps):
  train_loss, train_err = train_step()
  if step % num_steps_per_val == 0:
    val_loss, val_err = val()
    if val_err < best_val_err:
      best_val_err = val_err
      print('Step {:5d}: Obtained a best validation error of {:.3f}.'.format(step, best_val_err))
    info.append([step, train_loss, val_loss, train_err, val_err])
    x, y11, y12, y21, y22 = zip(*info)
    ax[0].plot(x, y11, x, y12)
    ax[0].legend(['Train loss', 'Val loss'])
    ax[1].plot(x, y21, x, y22)
    ax[1].legend(['Train err', 'Val err'])
```
